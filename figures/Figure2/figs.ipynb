```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
plt.rcParams['svg.fonttype'] = 'none'
```

```python
subsystems = pd.read_csv('subsystems.csv',sep='\t',index_col=0)
subsystems.head()
```

```python
subsystems['total of reactions'].sort_values(ascending=False).plot.pie(autopct='%1.0f%%')
plt.savefig('subsystems_pie.svg',format='SVG')
```

```python
progress = pd.read_csv('progress.csv',sep='\t',index_col=0)
progress
```

```python
progress.div(progress.max()).T.plot.bar(legend=False)
plt.savefig('progress.svg',format='SVG')
```
